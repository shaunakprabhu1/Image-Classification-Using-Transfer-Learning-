A CNN usually has an input layer, hidden layers, and an output layer or fully connnected layer. Each layer normally consist of:


1.   ConvNet: The convolutional mask that extracts features from the images
2.   Max-Pool: Used to reduce dimesionality of the image by reducing the feature map
3.   Flatten:Used to convert the data into a one dimensional array.
4.   The multi layer perceptron: Dense layer usually consisting of the number of output classes
5.   Dropout and Batch Normalization: Regularization layers to ensure that the model does not overfit.

The layers in a CNN are decided by the user and usually involves a fair degree of trial and error. Some hyperparameters in these models are usually kept consistent across most models.



## 3.Transfer Learning

Sometimes its much more convenient to use a pre-trained model which has been trained on a similar dataset than to use your own CNN and start from scratch. Transfer Learning uses the knowledge gained from having peformed one task to perform a different task. The key motivation for Transfer Learning, in the context of deep learning is the fact that most models which solve complex problems need a whole lot of data, and getting vast amounts of labeled data for supervised models can be really difficult, considering the time and effort it takes to label data points.

Transfer learning has the benefit of decreasing the training time for a neural network model and can result in lower generalization error which arises when a CNN overfits causing the model to have a low bias.
![alt text](https://drive.google.com/uc?id=1lPGpdzI-0NChnPdamDN0Wf-NXmZwxgPQ)

    https://towardsdatascience.com%2Fa-comprehensive-hands-on-guide-to-transfer-learning-with-real-world-applications-in-deep-learning-212bf3b2f27a&psig=AOvVaw3aOyIdbaHZ7y0bnoqDY7rV&ust=1587860821545000&source=images&cd=vfe&ved=0CA0QjhxqFwoTCPC92NSogukCFQAAAAAdAAAAABAD

There are many ways to implement a pre-trained model for your dataset. Popular models such as VGG, Xception, Inception V3 and Resnet50 are available for use but can be implemented in different ways to get the best results for your dataset. 

When fine tuning a pre-trained model you should keep the train-freeze tradeoff in mind. This deals with knowing which deciding which layers to freeze and which to train.

You can fine tune your pre-trained model for your particular dataset according one of three main strategies:


1.   Train the entire model: This would require a large dataset as in this case we are not going to be using any of the pre-loaded weights as we are re-training all of the layers. This would give a high accuracy but would require a lot of computation and time.
2.   Train only some layers: Here we can choose how much we want to adjust the weights of the model by freezing the top layers and training the lower layers.
3.  Training only the classifier: This is used for smaller datasets that have a little bit in common with the pre-trained models dataset. This technique invilves freezing all the convolutional layers and leaving only the output layer to specify the number of output nodes in your particular dataset

The image width, height, and batch size are initialized. 
Usually a batch size of 32, 64,or 128 is selected depending on the size of the dataset.

```python
img_height = 224
img_width = 224
data_list = os.listdir(train_data_dir)
num_classes = len(data_list)
batch_size = 64
categories = ['n0','n1','n2','n3','n4','n5','n6','n7','n8','n9']
```

Next we print a few images from each of the categories just to get an idea of how our dataset looks.

```python
for k in range(3):
    i=0
    plt.figure(figsize=(25,15))
    for category in categories:
        plt.subplot(5, 5, i+1)
        plt.yticks([])
        plt.xticks([])
        path=train_data_dir + '/' + category
        image_p=os.listdir(path)
        plt.title(category , color='red').set_size(15)
        plt.axis('off')
        image = cv2.imread(os.path.join(path, image_p[k])) 
        image = image[:, :, [2, 1, 0]] 
        plt.imshow(image)
        i+=1
```

```python
datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_generator = datagen.flow_from_directory(directory=train_data_dir, target_size=(img_width,img_height), color_mode='rgb',
                                              classes =categories, class_mode='categorical', batch_size=batch_size, shuffle = True)

```

```python
validation_generator =  datagen.flow_from_directory(directory=valid_data_dir, target_size=(img_width,img_height), color_mode='rgb',
                                              classes = categories, class_mode='categorical', batch_size=batch_size, shuffle = True)
```

```python
train_num = train_generator.samples
validation_num = validation_generator.samples
```

```python
train_generator.image_shape
```

```python
model = Sequential()

model.add(Conv2D(32,(3,3),input_shape = (224,224,3),activation='relu', kernel_initializer = 'random_uniform'))
  #model.add(BatchNormalization( momentum=0.85))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(32,(3,3),activation = 'relu'))
model.add(BatchNormalization( momentum=0.85))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))

model.add(Conv2D(64,(3,3),activation='relu', padding='same'))
model.add(Conv2D(64,(3,3),activation = 'relu'))
model.add(BatchNormalization( momentum=0.85))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(512,activation= 'relu', kernel_initializer='random_uniform'))
model.add(Dense(128, activation= 'relu'))
model.add(Dropout(0.2))
model.add(Dense(10,activation ='softmax'))
  
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=4)
  #This function is used to compile the model and selct an apporopriate gradient estimator and a cost function
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
#The model is trained and the number of epochs is selected
  his = model.fit_generator(train_generator, steps_per_epoch=train_num // batch_size, epochs= 50,
                            verbose = 1, validation_data = validation_generator, validation_steps = validation_num // batch_size )
 
 # plotting the train vs test acccuracy with respect to the number of epochs
  acc = his.history['accuracy']
  val_acc = his.history['val_accuracy']
  loss = his.history['loss']
  val_loss = his.history['val_loss']
  epochs = range(1, len(acc) + 1)

  plt.title('Training and validation accuracy')
  plt.plot(epochs, acc, 'red', label='Training acc')
  plt.plot(epochs, val_acc, 'blue', label='Validation acc')
  plt.legend()

  plt.figure()
  plt.title('Training and validation loss')
  plt.plot(epochs, loss, 'red', label='Training loss')
  plt.plot(epochs, val_loss, 'blue', label='Validation loss')

  plt.legend()

  plt.show()
```

From the graph we can see that the model tends to overfit the data after roughly 20 epochs.
This can be proven by looking at the how the validation loss does not further decrease after a point while the training loss continues to decrease.

Ultimately the model didnt too well giving a best validation accuracy of 64%

## ii) Creating VGG model

The VGG model was created using the imagenet dataset and has been trained over 1.2 million images to classify a thousand objects. The VGG16 architecture

instead of having a large number of hyper-parameter it focuses on having convolution layers of 3x3 filter with a stride 1 and uses same padding and maxpool layer of 2x2 filter of stride 2. It follows this arrangement of convolution and max pool layers consistently throughout the whole architecture. In the end it has 2 FC(fully connected layers) followed by a softmax for output. The 16 in VGG16 refers to the 16 layers that have weights.

![alt text](https://drive.google.com/uc?id=1WZcflvt_Sb1P0sbTSoAZVtKBcDYU0oqv)

    https://neurohive.io/en/popular-networks/vgg16/


In our case we will be training only the classifier layer of our VGG model and freezing all the other layers as the model is recognising monkey species which is similar to some of the categories it is already capable of recognising. Our dataset is also small ( <1000 images per class)

We start off with defining our VGG model and instantiating it.

```python
vgg16_model = keras.applications.vgg16.VGG16(weights='imagenet')
vgg16_model.summary()
```

Here we can see all the layers present in the VGG16 model. We can see the output layer has 1000 classes which we will need to change to 10 in our case

```python
!pip install gast==0.2.2
```

I was getting a bit of an error with VGG so had to downgrade the version of gast to 0.2.2. This will only need to be run if while running the model later you too get an error.

```python
model1 = Sequential()
for layer in vgg16_model.layers:
  model1.add(layer)
```

Here we create a sequential model and add all the layers of VGG16 to it. Below we get the summary for our sequential model

```python
model1.summary()
```

We remove the bottom layer of the model .e the classifier and add all the remaining layers.

```python
model1 = Sequential()
for layer in vgg16_model.layers[:-1]:
    model1.add(layer)
```

Next we freeze these layers so that they cant be trained. 

```python
for layer in model1.layers:
  layer.trainable = False
```

We add our classifier to the model with the softmax activation function

```python

model1.add(Dense(10, activation="softmax"))
model1.summary() # Finally our activation layer! we use 10 outputs as we have 10 monkey species (labels)
```

```python
es = EarlyStopping(patience=10, restore_best_weights=True)

# checkpoint to save model
chkpt = ModelCheckpoint(filepath="model1", save_best_only=True)
model1.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
#number of training and validation steps for training and validation
nb_train_steps = train_num // batch_size
nb_valid_steps = validation_num // batch_size

# number of epochs 
```

```python
his1 = model1.fit_generator(train_generator, 
                              epochs=50, 
                              steps_per_epoch=nb_train_steps, 
                              validation_data=validation_generator,
                              validation_steps=nb_valid_steps,
                              callbacks=[es,chkpt])

# plotting the train vs test acccuracy with respect to the number of epochs
acc = his1.history['accuracy']
val_acc = his1.history['val_accuracy']
loss = his1.history['loss']
val_loss = his1.history['val_loss']
epochs = range(1, len(acc) + 1)

plt.title('Training and validation accuracy')
plt.plot(epochs, acc, 'red', label='Training acc')
plt.plot(epochs, val_acc, 'blue', label='Validation acc')
plt.legend()

plt.figure()
plt.title('Training and validation loss')
plt.plot(epochs, loss, 'red', label='Training loss')
plt.plot(epochs, val_loss, 'blue', label='Validation loss')

plt.legend()

plt.show()
```

The VGG model performs substantially better than the CNN model we designed. Here we get a max accuracy on the validation set of 73%.
The validation loss also decreases proportinonally with the training loss thus indicating there isnt much overfitting.

### iii) creating the XCEPTION model

This model was developed by François Chollet in 2017 and overtook Inception model in both speed and accuracy.
The Xception model is very different from the VGG16 in a sense it is not execute sequentially in keras
![alt text](https://drive.google.com/uc?id=1wk1H-19-lvG0lZxuMWX6IPz5rlb76CzO)

    https://www.intersysconsulting.com/blog/convolutional-neural-network-2/

The architecture for the Xception model is based on depthwise separable convolution layers and consists of three major sections: Entry Flow, Middle Flow, and Exit Flow.

Xception is an efficient architecture that works by using Depthwise Seperable convolutios and shortcuts between the convolution blocks.

This is an extremely deep model and is very expensive to train if starting from scratch. 
In our applicatin we will only be replacing the outplut MLP layer with the number of classes we need.

The model is already available in keras so it makes working with it very easy. 

We start of by instantiating it and loading the model in a variable 'xception'

```python
xception = keras.applications.Xception(weights='imagenet', include_top=False)
xception.summary()
```

The summary function gives an idea of the number of parameters in the model and the number of layers.

The xception model is a bit rigid when it comes to the number of parameters it takes as input. We therefore change the input image size to 299 by 299 to make it compatible. 

We also change the batch size to 4 to make the layers compatible when executing.

```python
train_generator = datagen.flow_from_directory(directory=train_data_dir, target_size=(299,299), color_mode='rgb',
                                              classes =categories, class_mode='categorical', batch_size=4, shuffle = True)

validation_generator =  datagen.flow_from_directory(directory=valid_data_dir, target_size=(299,299), color_mode='rgb',
                                              classes = categories, class_mode='categorical', batch_size=4, shuffle = True)

train_num = train_generator.samples
validation_num = validation_generator.samples
```

We freeze all the layers making them untrainable.

```python
for layer in xception.layers:
  layer.trainable = False
```

The Xception architecture requires using a global average pooling layer before the output layer. 
We use a branch method here and not sequential.

```python
x = xception.output
x = GlobalAveragePooling2D()(x)
#x = Dense(512,activation= 'relu')(x)
output = Dense(10, activation='softmax')(x)
model2 = Model(xception.input,output)
model2.summary()
```

```python
# es = EarlyStopping(patience=10, restore_best_weights=True)

# # checkpoint to save model
# chkpt = ModelCheckpoint(filepath="model1", save_best_only=True)
model2.compile(optimizer=SGD(lr=1e-4, momentum=0.9), loss='categorical_crossentropy', metrics=['accuracy'])
#number of training and validation steps for training and validation
nb_train_steps = train_num // batch_size
nb_valid_steps = validation_num // batch_size
```

```python
#with tf.device("/device:GPU:0"):
his2= model2.fit_generator(train_generator, 
                                epochs=10, 
                                #steps_per_epoch=nb_train_steps, 
                                validation_data=validation_generator,verbose = 1,use_multiprocessing=True)
                                #validation_steps=nb_valid_steps)
                                #callbacks=[es,chkpt])

# plotting the train vs test acccuracy with respect to the number of epochs
acc = his2.history['accuracy']
val_acc = his2.history['val_accuracy']
loss = his2.history['loss']
val_loss = his2.history['val_loss']
epochs = range(1, len(acc) + 1)

plt.title('Training and validation accuracy')
plt.plot(epochs, acc, 'red', label='Training acc')
plt.plot(epochs, val_acc, 'blue', label='Validation acc')
plt.legend()

plt.figure()
plt.title('Training and validation loss')
plt.plot(epochs, loss, 'red', label='Training loss')
plt.plot(epochs, val_loss, 'blue', label='Validation loss')

plt.legend()

plt.show()
```
